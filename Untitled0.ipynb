```python
 import requests as rq
```

```python
def data_collection(title) :
  params = { 
        'action': 'query', 
        'format': 'json', # request json formatted content
        'titles': title, # title of the wikipedia page
        'prop': 'extracts', 
        'explaintext': True
    }
  r=rq.get('https://en.wikipedia.org/w/api.php',params=params).json()
  ##Extraire les donnees du dictionnaire
  page=next(iter(r['query']['pages'].values()))
  if 'extract' in page.keys():
    return page['extract']
  else:
    return "Page not found"
txt=data_collection('Data science').lower()
print(txt)
```

```python

# this is the url for Alice in Wonderland
result = rq.get('http://www.gutenberg.org/files/11/11-0.txt')
print(result.text)
```

# *The definition of stop words ❓*

Stp words :are words that do not provide any useful information to infer content or nature. It may be either because they don't carry any meaning (prepositions, conjunctions, etc.) or because they are too frequent.

Counting word frequencies is a three-step process:

  The first step is to create a list of the most frequent words by splitting the text over the whitespace character ' ' with the function text.split(' ').

  Then, count how many times each word appears in the text using the  Counter  function.




```python
from collections import Counter
word_list=txt.split(' ')
word_counts=Counter(word_list)
##The word_counts is a dictionnary .keys=>words,values=>accurences
```

```python
print(word_counts)
```

```python
for i in iter(word_counts):
  if word_counts[i] >10 :
    print(i,word_counts[i])
##Second code :
# for w in word_counts.most_common(10):
#   print(f"{w[0]} : {w[1]} ")
```

The problem with this method is that some words are meaningful as data ,science information statistics ....
The second alternative is to create a list of remove words

```python
stop_words=['is','that','and','to','from','a','of','in','the','as','for','used']
```

```python
wordcloud1=WordCloud(width=400,height=200,stopwords=stop_words)
wordcloud1.generate(txt)
plt.imshow(wordcloud1)
```

# **Tokenization**
Transforming a text into a list of words is called tokenization, and each word is a token.


In the previous section ,we used the function split for tokenization ,is it  the best method ??
No ,why ??
 

1.   Splitting on whitespaces does not take into account punctuation or contractions
2.   Always keep in mind that the size of the vocabulary directly impacts the necessary computing power. 

and does not scale.

***As a general rule, good tokenization should handle*** 

*   Punctuation:  eat, => ["eat", ","]
*   Contractions:
> can't => ["can", "'t"] or ["can", "'", "t"] ; doesn't => ["doesn", "'t"] or ["doesn", "'", "t"]
